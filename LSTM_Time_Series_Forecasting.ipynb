## LSTM Model

- Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
dataset = temp.to_numpy().reshape(-1, 1)
train_size = 60000
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(dataset[:train_size])
scaled_dataset = scaler.transform(dataset)
```

- Sequencing the data

```python
def df_to_X_y(df, window_size=5):
  df_as_np = df
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = [[a] for a in df_as_np[i:i+window_size]]
    X.append(row)
    label = df_as_np[i+window_size]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X1, y1 = df_to_X_y(scaled_dataset, WINDOW_SIZE)
X1.shape, y1.shape
```

- Splitting

```python
X_train1, y_train1 = X1[:60000], y1[:60000]
X_val1, y_val1 = X1[60000:65000], y1[60000:65000]
X_test1, y_test1 = X1[65000:], y1[65000:]
X_train1.shape, y_train1.shape, X_val1.shape, y_val1.shape, X_test1.shape, y_test1.shape
```

- Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model1 = Sequential()
model1.add(InputLayer((5, 1)))
model1.add(LSTM(64))
model1.add(Dense(8, 'relu'))
model1.add(Dense(1, 'linear'))

model1.summary()
```

```python
cp1 = ModelCheckpoint('model1/', save_best_only=True)
model1.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001), metrics=[RootMeanSquaredError()])
```

```python
model1.fit(X_train1, y_train1, validation_data=(X_val1, y_val1), epochs=10, callbacks=[cp1])
```

```python
from tensorflow.keras.models import load_model
model1 = load_model('model1/')
```

- Predictions on train and val data

```python
scaledtrain_predictions = model1.predict(X_train1)
train_predictions = scaler.inverse_transform(scaledtrain_predictions)
y_train1 = scaler.inverse_transform(y_train1.reshape(-1, 1))
```

```python
train_results = pd.DataFrame(data={'Train Predictions':train_predictions.flatten(), 'Actuals':y_train1.flatten()})
train_results
```

```python
import matplotlib.pyplot as plt
plt.plot(train_results['Train Predictions'][50:100])
plt.plot(train_results['Actuals'][50:100])
```

```python
scaledval_predictions = model1.predict(X_val1)
val_predictions = scaler.inverse_transform(scaledval_predictions)
y_val1 = scaler.inverse_transform(y_val1.reshape(-1, 1))
```

```python
val_results = pd.DataFrame(data={'Val Predictions':val_predictions.flatten(), 'Actuals':y_val1.flatten()})
val_results
```

```python
plt.plot(val_results['Val Predictions'][:100])
plt.plot(val_results['Actuals'][:100])
```

- Test Predictions

```python
scaledtest_predictions = model1.predict(X_test1)
test_predictions = scaler.inverse_transform(scaledtest_predictions)
y_test1 = scaler.inverse_transform(y_test1.reshape(-1, 1))
```

```python
test_results = pd.DataFrame(data={'Test Predictions':test_predictions.flatten(), 'Actuals':y_test1.flatten()})
test_results
```

```python
plt.plot(test_results['Test Predictions'][:100])
plt.plot(test_results['Actuals'][:100])
```

- Evaluation

```python
MAE_LSTM = mean_absolute_error(test_results['Test Predictions'],test_results['Actuals'])
```
